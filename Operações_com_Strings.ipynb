```python
#Como juntar o separador de milhar.
formato_US="67231286532.541"
#Colocando como lista as partes separadas pelo ponto
formato_US=formato_US.split(".")
#Parte inteira é o primeiro elemento da lista
parte_inteira=formato_US[0]
#Parte decimal é o segundo elemento da lista
parte_decimal=formato_US[1]
#Calcular o resto da divisão por 3 do tamanho da parte inteira
resto=len(parte_inteira)%3
#Se o tamanho da parte inteira não for múltipla de 3
if resto!=0:
    #Calcula o quociente da divisão do tamanho da parte inteira por 3
    blocos_inteira=len(parte_inteira)//3
    #Armazena os 1 ou 2 primeiros dígitos em numero_convertido
    numero_convertido=[parte_inteira[0:resto]]
    #Adiciona na lista os outros blocos de números, 3 a 3, usando slice.
    for i in range(blocos_inteira):
        numero_convertido.append(parte_inteira[(resto+3*i):(resto+3*(i+1))])
    #Junta os elementos da lista com ponto
    numero_convertido=".".join(numero_convertido)
    #Junta a parte convertida e a parte decimal com a vírgula
    numero_convertido=f"{numero_convertido},{parte_decimal}"
else:
    blocos_inteira=len(parte_inteira)//3
    numero_convertido=[]
    for i in range(blocos_inteira):
        numero_convertido.append(parte_inteira[(resto+3*i):(resto+3*(i+1))])
    numero_convertido=".".join(numero_convertido)
    numero_convertido=f"{numero_convertido},{parte_decimal}"
print(numero_convertido)
```

# Expressões regulares (RegEx)

```python
import re

# Exemplo de string com dígitos e outros caracteres
#texto_exemplo = "O CPF 023.456.789-00 possui 11 dígitos, mas o número de telefone é (11) 98765-4321."

# A expressão regular r'\d+' encontra uma ou mais ocorrências de dígitos.
# re.findall retorna uma lista de todas as ocorrências.
##digitos_encontrados = re.findall(r'\d+', texto_exemplo)

# Para juntar todos os dígitos em uma única string, usamos o método .join()
#somente_digitos = "".join(digitos_encontrados)

#print(f"Texto original: {texto_exemplo}")
#print(f"Dígitos encontrados (lista): {digitos_encontrados}")
#print(f"Somente dígitos (string): {somente_digitos}")

# Outro exemplo com o CPF do kernel
cpf_com_formatacao = "023*456*789*0"
cpf_somente_digitos = "".join(re.findall(r'\d+', cpf_com_formatacao))
print(f"\nCPF original: {cpf_com_formatacao}")
print(f"CPF com somente dígitos: {cpf_somente_digitos}")
print(len(cpf_somente_digitos))
```

```python
cpf_com_formatacao = "023.456.789.00"
cpf_somente_digitos = "".join(re.findall(r'\d+', cpf_com_formatacao))
print(cpf_somente_digitos)
```

```python
#Função para determinar os dígitos verificadores
def cpf(lista):
    #Calculo para o primeiro digito
    soma_digito1=0
    for i in range(len(lista)):
        soma_digito1 += lista[i]*(10-i)
    if soma_digito1 % 11 == 0 or soma_digito1 % 11 == 0:
        digito1=0
    else:
        digito1=(11-(soma_digito1%11))
    lista.append(digito1)
    #Cálculo para segundo dígito
    soma_digito2=0
    for i in range(len(lista)):
        soma_digito2+=lista[i]*(11-i)
    if soma_digito2 % 11 == 0 or soma_digito2 % 11 == 0:
        digito2=0
    else:
        digito2=(11-(soma_digito2%11))
    lista.append(digito2)
    return lista
```

```python
#teste=[1,3,3,4,8,8,5,4,4]
teste=[0,4,7,3,3,4,3,8,4]
meucpf=cpf(teste)
meucpf=str(meucpf)
meucpf=meucpf.replace("[","")
meucpf=meucpf.replace("]","")
meucpf=meucpf.replace(",","")
meucpf=meucpf.replace(" ","")
print(meucpf)
verificadores=meucpf[9:11]
meucpf=[meucpf[0:3],meucpf[3:6],
        meucpf[6:9]]
meucpf=".".join(meucpf)
meucpf=f"{meucpf}-{verificadores}"
print(meucpf)
```
